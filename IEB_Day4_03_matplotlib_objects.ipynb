## Plotting with matplotlib objects

### What you will learn in this session

- How to initiate a figure and axes object
- Plotting using the axes object
- Customizing using the axes object
- saving a plot

### Examples from the intro

#### Using the ax object to plot

```python
import yfinance as yf
import matplotlib.pyplot as plt

aapl = yf.download('AAPL', period='5y')

fig, ax = plt.subplots()

ax.plot(aapl['Adj Close']) # line plot using the .plot() method

plt.show()
```

#### Using the ax object to set plot properties like the title

```python
aapl = yf.download('AAPL', period='5y')

fig, ax = plt.subplots()

ax.plot(aapl['Adj Close'])

ax.set_title('Apple Adj Close') # setting title with set_title() method

plt.show()
```

#### saving a figure using the fig object

```python
fig.savefig('aapl_adj.png')
```
