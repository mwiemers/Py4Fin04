## Matplotlib - creating subplots

What you will learn in this unit:

### Examples from the intro

#### creating subplots

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
```

You can also unpack the list into separate variables.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
```

#### Plotting to different subplots

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(aapl['Adj Close']) # using .plot() to plot the adjusted price in the first plot
ax2.hist(aapl['Adj Close'].pct_change()) # using .hist() to plot the histogram of the return rates in second plot

plt.show()
```

#### Setting properties for different subplots

```python
aapl = yf.download('AAPL', period='5y')

fig, ax = plt.subplots(nrows=2, ncols=1)

#first plot
ax[0].plot(aapl['Adj Close']) # plot adjusted close data as line
ax[0].set_ylabel('Adj Close in USD') # set the ylabel

ax[1].hist(aapl['Adj Close'].pct_change(), bins=20) # plot histogram of adjusted close data
ax[1].set_xlabel('Return rates') #  set the xlabel

plt.show()
```
